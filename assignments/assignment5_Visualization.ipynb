# Visualization

```python
import pandas as pd
history = pd.read_csv('history.csv')
```

```python
history
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Train Loss  Train Accuracy
axes[0].plot(history[history['Model'] == 'BasicCNN']['Epoch'], history[history['Model'] == 'BasicCNN']['Train Loss'], label='BasicCNN Train Loss', color='tab:red')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Train Loss')
axes[0].set_title('Train Loss and Acc')
axes[0].legend(loc='upper left')

ax2 = axes[0].twinx()
ax2.plot(history[history['Model'] == 'BasicCNN']['Epoch'], history[history['Model'] == 'BasicCNN']['Train Acc'], label='BasicCNN Train Acc', linestyle='--', color='tab:orange')
ax2.set_ylabel('Train Accuracy')
ax2.legend(loc='lower right')

axes[1].plot(history[history['Model'] == 'BasicCNN']['Epoch'], history[history['Model'] == 'BasicCNN']['Val Loss'], label='BasicCNN Val Loss', color='tab:red')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Val Loss')
axes[1].set_title('Validation Loss and Acc')
axes[1].legend(loc='upper left')

ax2 = axes[1].twinx()
ax2.plot(history[history['Model'] == 'BasicCNN']['Epoch'], history[history['Model'] == 'BasicCNN']['Val Acc'], label='BasicCNN Val Acc', linestyle='--', color='tab:orange')
ax2.set_ylabel('Val Accuracy')
ax2.legend(loc='lower right')

axes[2].plot(history[history['Model'] == 'BasicCNN']['Epoch'], history[history['Model'] == 'BasicCNN']['Epoch Time'], label='BasicCNN Epoch Time', color='tab:red')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('Epoch Time (seconds)')
axes[2].set_title('Epoch Time')
axes[2].set_ylim(0, 20)
axes[2].legend()

plt.tight_layout()
plt.show()
```
